```python
books['Book-Title'] = books['Book-Title'].apply(lambda name : name.lower())
```

popularity based recommendation system means highest or threshold average ratings books should be shown to users

```python
# merge_df_rating_book
```

```python
merge_df_rating_book = ratings.merge(books, on='ISBN')
```

```python
rating_count_df = merge_df_rating_book.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
rating_count_df = rating_count_df.rename(columns={'Book-Rating':'rating_count'})
rating_count_df
```

```python
max(rating_count_df['rating_count']) 
```

```python
rating_count_df[rating_count_df['rating_count'] == 2502]
```

```python
avg_rating_df = merge_df_rating_book.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df = avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'})
avg_rating_df
```

```python
popular_books_df = rating_count_df.merge(avg_rating_df,on='Book-Title')
popular_books_df
```

```python
popular_books_df = popular_books_df[popular_books_df.rating_count>=250]
```

```python
popular_books_df = popular_books_df.sort_values(by='avg_ratings',ascending=False)
```

```python
popular_books_df = popular_books_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','rating_count','avg_ratings']]
```

```python
def recommeder(book):

    index = np.where(pivot_df.index == book)[0][0]

    similar_books = sorted(list(enumerate(cosine_similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    book_names = []
    for i in similar_books:
        name = pivot_df.index[i[0]]
        book_names.append(name)
    return book_names
```

```python
recommeder('1984')
```

```python
def recommend(book_name):
    index = np.where(pivot_df.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(cosine_similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pivot_df.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
data = recommend('1984')
```
